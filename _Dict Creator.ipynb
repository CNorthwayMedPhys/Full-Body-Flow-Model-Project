```python
import pandas as pd
import numpy as np
```

```python
#Load in the arterial map file
try:
    vessel_df = pd.read_pickle ('C:\\Users\\cbnor\\Documents\\Full Body Flow Model Project\\SysArteries.pkl')
except:
    vessel_df = pd.read_pickle ('C:\\Users\\Cassidy.Northway\\RemoteGit\\SysArteries.pkl')
```

```python
#Note the index number of each segment which ends in a ST in a new array

filtered_df = vessel_df[vessel_df['End Condition'] == 'ST']
np_array = filtered_df.index.values
```

```python
#Create a second column with the matching segment numbers based on the matching map
empty_col = np.zeros((np.shape(np_array)[0],1))
array = np.append(np.expand_dims(np_array,axis=1),empty_col,axis=1)

#Read in the matching vessel names 
try:
    mirror_df = pd.read_excel ('C:\\Users\\cbnor\\Documents\\Full Body Flow Model Project\\Mirroring_Tuning_Map.xlsx', usecols ="B,C,E")
except:
    mirror_df = pd.read_excel ('C:\\Users\\Cassidy.Northway\\RemoteGit\\Mirroring_Tuning_Map.xlsx', usecols ="B,C,E")

left_vessels = mirror_df['Left Vessel'].reset_index()
right_vessels = mirror_df['Right Vessel'].dropna().reset_index()

#Search the vessel data frame to find string containing the vessel names, match to the index and update the second column
for i in range(0,len(right_vessels)):
    left_string = left_vessels['Left Vessel'].iloc[i]
    index_left = filtered_df[filtered_df['Name'].str.match(left_string+'_')].index.values

    right_string = right_vessels['Right Vessel'].iloc[i]
    index_right = filtered_df[filtered_df['Name'].str.match(right_string+'_')].index.values

    array_index = np.where(array[:,0] == index_left)
    array[array_index,1] = index_right
    
    #Remove the matched value rows
    rem_index = np.where(array[:,0] == index_right)
    array = np.delete(array,rem_index, axis = 0)

np_array = array
```
