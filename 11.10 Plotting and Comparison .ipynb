```python
import numpy as np
import matplotlib.pylab as plt
from scipy import interpolate
import tkinter as tk
from tkinter import filedialog
```

```python
#Select files 
root = tk.Tk()
root.withdraw()

# open the file dialog box and select data file
file_path = filedialog.askopenfilename()
file_namea = file_path.split('/')[-1]
file_namea = file_namea.split('.')[0]
```

```python
#Makes the plot look chefs kiss

WIDTH = 510  # the number latex spits out
FACTOR = 1.0  # the fraction of the width you'd like the figure to occupy
fig_width_pt  = WIDTH * FACTOR
inches_per_pt = 1.0 / 72.27
golden_ratio  = (np.sqrt(5) - 1.0) / 2.0  # because it looks good
fig_width_in  = fig_width_pt * inches_per_pt  # figure width in inches
fig_height_in = fig_width_in * golden_ratio   # figure height in inches
fig_dims    = [fig_width_in, fig_height_in] # fig dims as a list
```

```python
data_a = np.loadtxt(file_path, delimiter = ',') #3d in time and space
x = np.arange(0,np.size(data_a[:,0]),1)

plt.figure(figsize=fig_dims)
plt.plot(data_a[:,0], label = "Proximal", c = 'darkorchid', ls = 'dotted')
plt.plot(data_a[:,-1], label = "Distal", c= 'darkorchid')
plt.legend()
plt.title(file_namea)
plt.xlabel('t (s)')
plt.ylabel('flow rate (cm^3/s)')
plt.xlim([min(x), max(x)])
```

```python
#Select files 
root = tk.Tk()
root.withdraw()

# open the file dialog box and select data file
file_path = filedialog.askopenfilename()
file_nameb = file_path.split('/')[-1]
file_nameb = file_nameb.split('.')[0]
```

```python
data_b = np.loadtxt(file_path, delimiter = ',') #3d in time and space

plt.figure(figsize=fig_dims)
plt.plot(data_b[:,0], label = "Proximal", c = 'darkorchid', ls = 'dotted')
plt.plot(data_b[:,-1], label = "Distal", c= 'darkorchid')
plt.legend()
plt.title(file_nameb)
plt.xlabel('time ')
plt.ylabel('flow rate (cm^3/s)')
plt.xlim([min(x), max(x)])
```

```python
#Difference between the values

diff = data_b - data_a


plt.figure(figsize=fig_dims)
plt.plot(diff[:,0], label = "Proximal", c = 'darkorchid', ls = 'dotted')
plt.plot(diff[:,-1], label = "Distal", c= 'darkorchid')
plt.legend()
plt.title('Difference between Data Sets')
plt.xlabel('time ')
plt.ylabel('flow rate (cm^3/s)')
plt.xlim([min(x), max(x)])
```
