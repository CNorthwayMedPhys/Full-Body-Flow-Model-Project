```python
#Import
import pandas as pd
import numpy as np
```

```python
vessel_df = pd.read_pickle ('C:\\Users\\Cassidy.Northway\\RemoteGit\\SysArteries.pkl')
display(vessel_df)
```

```python
vessel_df = vessel_df.filter(items = ['Name','End Condition'])
lrr_df = vessel_df.loc[vessel_df['End Condition'] == 'ST']

lrr_df['Orginal'] = 'Yes'
display(lrr_df)
```

```python
#Load in mirroring document
data = pd.read_excel('C:\\Users\\Cassidy.Northway\\RemoteGit\\Mirroring_Tuning_Map.xlsx', sheet_name = 0)
mirror_df = data.dropna(subset=["Right Vessel"])
display(mirror_df)
```

```python
for ind in mirror_df.index:
    mirrored_vessel = mirror_df.loc[ind,'Right Vessel']+'_'
    orginal_vessel = mirror_df.loc[ind,'Left Vessel']+"_"
    
    orginal_index = lrr_df[lrr_df['Name'].str.startswith(orginal_vessel)].index[0]
    mirrored_index = lrr_df[lrr_df['Name'].str.startswith(mirrored_vessel)].index[0]
    
    lrr_df.loc[mirrored_index, 'Orginal'] = str(orginal_index)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):   
    display(lrr_df)
```
