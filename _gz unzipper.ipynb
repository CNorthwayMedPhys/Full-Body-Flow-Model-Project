```python

import gzip

with gzip.open('data.tar.gz','r') as fin:        
    for line in fin:        
        print('got line', line)
```

```python
import gzip
import shutil
 
with gzip.open('data.tar.gz', 'rb') as f_in:
  with open('data.txt', 'wb') as f_out:
    shutil.copyfileobj(f_in, f_out)
```
