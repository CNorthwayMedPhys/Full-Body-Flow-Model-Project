```python
#!pip install Cmake
#!pip install wheel
#!pip install openmesh
!pip install openmesh --no-cache-dir
#%pip install trimesh

import openmesh as om
import numpy as np
import trimesh as tm
import tkinter
from tkinter import filedialog

tkinter.Tk().withdraw() # prevents an empty tkinter window from appearing
file_names = filedialog.askopenfilenames()
```

```python
vessels = 0 #Turn on (=1) if converting vessels (or want sorting into veins and arteries)
whole = 0 #Turn on (=1) if you want to just plot the mesh as a single entity
components = 1 #Turn on (=1) if you want to create a seperate file for 
```

```python
for file_name in file_names:
    with open(file_name) as file:
        file_contents = file.read() #import as string
        lines = file_contents.splitlines() #break into list by lines 
    if vessels == 1 or whole == 1: 
        lines = lines[1:]
        triangles_art = [] #for arteries
        triangles_veins = [] # for veins
        triangles =[] #for anyother structures I want to convert
        for line in lines: #going line by line
            items = line.split(' ')
            if vessels == 1:
                mesh_art = om.TriMesh() #Empty mesh
                mesh_veins = om.TriMesh()
                if 'art' in items or 'arteries' in items or 'veins' in items:
                    if 'veins' in items:
                        tag = 0
                    else:
                        tag = 1
                else:              
                    try:                     #removes all empty spaces from list
                        while True:
                            items.remove('')
                            try:
                                items.remove('') #sometimes we have double gaps
                            except:
                                pass
                    except ValueError:
                        pass
                    if tag == 0:
                        triangles_veins.append(items)
                    else:
                        triangles_art.append(items)
            if whole == 1:
                mesh = om.TriMesh()
                try:                     #removes all empty spaces from list
                    while True:
                        items.remove('')
                        try:
                            items.remove('') #sometimes we have double gaps
                        except:
                            pass
                        triangles.append(items)
                except ValueError:
                    pass

             #Here is where we create the mesh

        if triangles_art:
            for line in triangles_art:
       #Create points which will be the vertices
                try:
                    p0 = [float(line[0]), float(line[1]), float(line[2])]
                    p1 = [float(line[3]), float(line[4]), float(line[5])]
                    p2 = [float(line[6]), float(line[7]), float(line[8])]
                except ValueError:
                    pass

        #create vertices
            vh0 = mesh_art.add_vertex(p0)
            vh1 = mesh_art.add_vertex(p1)
            vh2 = mesh_art.add_vertex(p2)

        #Create face
            f0 = mesh_art.add_face(vh0,vh1,vh2)


        #Write as .stl file
            file_name_list=file_name.split("/")
            file_name_title=file_name_list[-1].split(".")
            om.write_mesh(file_name_title[0] + 'art.stl',mesh_art) 

        if triangles_veins:
            for line in triangles_veins:

       #Create points which will be the vertices
                try:
                    p0 = [float(line[0]), float(line[1]), float(line[2])]
                    p1 = [float(line[3]), float(line[4]), float(line[5])]
                    p2 = [float(line[6]), float(line[7]), float(line[8])]
                except ValueError:
                    pass

        #create vertices
                vh0 = mesh_veins.add_vertex(p0)
                vh1 = mesh_veins.add_vertex(p1)
                vh2 = mesh_veins.add_vertex(p2)

        #Create face
                f0 = mesh_veins.add_face(vh0,vh1,vh2)

        #Write as .stl file
            file_name_list=file_name.split("/")
            file_name_title=file_name_list[-1].split(".")
            om.write_mesh(file_name_title[0] + 'veins.stl',mesh_veins)     
        if triangles:
            for line in triangles:

       #Create points which will be the vertices
                try:
                    p0 = [float(line[0]), float(line[1]), float(line[2])]
                    p1 = [float(line[3]), float(line[4]), float(line[5])]
                    p2 = [float(line[6]), float(line[7]), float(line[8])]
                except ValueError:
                    pass

        #create vertices
                vh0 = mesh.add_vertex(p0)
                vh1 = mesh.add_vertex(p1)
                vh2 = mesh.add_vertex(p2)

        #Create face
                f0 = mesh.add_face(vh0,vh1,vh2)

        #Write as .stl file
            file_name_list=file_name.split("/")
            file_name_title=file_name_list[-1].split(".")
            om.write_mesh(file_name_title[0] + '.stl',mesh)
            
#This will plot each individual component of the file as a unique mesh           
    if components == 1:
        lines = lines[1:]
        header = lines[0]
        triangles = []
        mesh = om.TriMesh() 
        i = 0 
        for line in lines[1:]: #going line by line
            items = line.split(' ')
            if any(c.isalpha() for c in items[0]): #If we hit a new header we need to write the existing mesh                 
                for row in triangles:
                    p0 = [float(row[0]), float(row[1]), float(row[2])]
                    p1 = [float(row[3]), float(row[4]), float(row[5])]
                    p2 = [float(row[6]), float(row[7]), float(row[8])]
                    vh0 = mesh.add_vertex(p0)
                    vh1 = mesh.add_vertex(p1)
                    vh2 = mesh.add_vertex(p2)
                    f0 = mesh.add_face(vh0,vh1,vh2)
                filename = header +str(i)
                om.write_mesh(header + '.stl',mesh) 
                mesh = om.TriMesh() #empty mesh
                triangles =[]
                if header == items [0]:
                    i = i + 1
                else:
                    i = 0
                header = items[0]
            else:

                items.remove('')
                try:
                    items.remove('') #sometimes we have double gaps
                except:
                    pass
                triangles.append(items)
        for row in triangles: #grabs the last one
            p0 = [float(row[0]), float(row[1]), float(row[2])]
            p1 = [float(row[3]), float(row[4]), float(row[5])]
            p2 = [float(row[6]), float(row[7]), float(row[8])]
            vh0 = mesh.add_vertex(p0)
            vh1 = mesh.add_vertex(p1)
            vh2 = mesh.add_vertex(p2)
            f0 = mesh.add_face(vh0,vh1,vh2)
        filename = header + str(i)
        om.write_mesh(header + '.stl',mesh) 
            
               
       
        

```
