```python
import numpy as np
import tkinter
from tkinter import filedialog
from scipy import interpolate
import matplotlib.pyplot as plt

T = 0.914
tc = 4
tf = T*tc
```

```python
#Import file

tkinter.Tk().withdraw() # prevents an empty tkinter window from appearing
file_name = filedialog.askopenfilename()
```

```python
U = np.loadtxt(file_name, delimiter=',')

L= U.shape[0]
t = np.linspace(tf-T, tf, U.shape[1])
x = np.linspace(0,L,U.shape[0])
g = interpolate.interp2d(t, x, U, kind='linear')
    
x = np.linspace(0, L, len(t))
U = g(t, x)
    
WIDTH = 510  # the number latex spits out
FACTOR = 1.0  # the fraction of the width you'd like the figure to occupy
fig_width_pt  = WIDTH * FACTOR
inches_per_pt = 1.0 / 72.27
golden_ratio  = (np.sqrt(5) - 1.0) / 2.0  # because it looks good
fig_width_in  = fig_width_pt * inches_per_pt  # figure width in inches
fig_height_in = fig_width_in * golden_ratio   # figure height in inches
fig_dims    = [fig_width_in, fig_height_in] # fig dims as a list
    
nt = len(t)
x = np.linspace(0, L, nt)
Y, X = np.meshgrid(x, t)
pt = U.shape[0]
dz = int(pt/nt)
Z_1 = U[0:pt:dz,:]


plt.figure(figsize=fig_dims)
plt.plot( X, Z_1[:,0], label = "Proximal", c = 'darkorchid', ls = 'dotted')
plt.plot(X, Z_1[:,-1], label = "Distal", c= 'darkorchid')

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())


plt.title(file_name)
plt.xlabel('t (s)')
plt.ylabel('flow rate (cm^3/s)')
plt.xlim([min(t), max(t)])


#fname = "%s/%s/%s_q3d%d.png" % (plot_dir, suffix, suffix, pos)
#ig.savefig(fname, dpi=600, bbox_inches='tight')
        


    
```
